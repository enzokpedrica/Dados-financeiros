```python
import pandas as pd
import os
from datetime import datetime


pasta_2021 = 'C:\\git\\Dados-financeiros\\01 - Bronze\\2021'
pasta_2022 = 'C:\\git\\Dados-financeiros\\01 - Bronze\\2022'
pasta_2023 = 'C:\\git\\Dados-financeiros\\01 - Bronze\\2023'
pasta_2024 = 'C:\\git\\Dados-financeiros\\01 - Bronze\\2024'



# Dicionário com nomes personalizados para os arquivos
arquivos_2021 = {
    'janeiro_21': '2021-01.csv',
    'fevereiro_21': '2021-02.csv',
    'marco_21': '2021-03.csv',
    'abril_21': '2021-04.csv',
    'maio_21': '2021-05.csv',
    'junho_21': '2021-06.csv',
    'julho_21': '2021-07.csv',
    'agosto_21': '2021-08.csv',
    'setembro_21': '2021-09.csv',
    'outubro_21': '2021-10.csv',
    'novembro_21': '2021-11.csv',
    'dezembro_21': '2021-12.csv'
}
arquivos_2022 = {
    'janeiro_22': '2022-01.csv',
    'fevereiro_22': '2022-02.csv',
    'marco_22': '2022-03.csv',
    'abril_22': '2022-04.csv',
    'maio_22': '2022-05.csv',
    'junho_22': '2022-06.csv',
    'julho_22': '2022-07.csv',
    'agosto_22': '2022-08.csv',
    'setembro_22': '2022-09.csv',
    'outubro_22': '2022-10.csv',
    'novembro_22': '2022-11.csv',
    'dezembro_22': '2022-12.csv'
}
arquivos_2023 = {
    'janeiro_23': '2023-01.csv',
    'fevereiro_23': '2023-02.csv',
    'marco_23': '2023-03.csv',
    'abril_23': '2023-04.csv',
    'maio_23': '2023-05.csv',
    'junho_23': '2023-06.csv',
    'julho_23': '2023-07.csv',
    'agosto_23': '2023-08.csv',
    'setembro_23': '2023-09.csv',
    'outubro_23': '2023-10.csv',
    'novembro_23': '2023-11.csv',
    'dezembro_23': '2023-12.csv'
}
arquivos_2024 = {
    'janeiro_24': '2024-01.csv',
    'fevereiro_24': '2024-02.csv',
    'marco_24': '2024-03.csv',
    'abril_24': '2024-04.csv',
    'maio_24': '2024-05.csv',
    'junho_24': '2024-06.csv',
    'julho_24': '2024-07.csv',
    'agosto_24': '2024-08.csv',
    'setembro_24': '2024-09.csv',
    'outubro_24': '2024-10.csv',
    'novembro_24': '2024-11.csv',
    'dezembro_24': '2024-12.csv'
}
dfs_21 = []
dfs_22 = []
dfs_23 = []
dfs_24 = []


for nome, arquivo in arquivos_2021.items():
    caminho_completo = os.path.join(pasta_2021, arquivo)
    df_21 = pd.read_csv(caminho_completo)
    df_21['arquivo_origem'] = nome
    dfs_21.append(df_21)

for nome, arquivo in arquivos_2022.items():
    caminho_completo = os.path.join(pasta_2022, arquivo)
    df_22 = pd.read_csv(caminho_completo)
    df_22['arquivo_origem'] = nome
    dfs_22.append(df_22)

for nome, arquivo in arquivos_2023.items():
    caminho_completo = os.path.join(pasta_2023, arquivo)
    df_23 = pd.read_csv(caminho_completo)
    df_23['arquivo_origem'] = nome
    dfs_23.append(df_23)

for nome, arquivo in arquivos_2024.items():
    caminho_completo = os.path.join(pasta_2024, arquivo)
    df_24 = pd.read_csv(caminho_completo)
    df_24['arquivo_origem'] = nome
    dfs_24.append(df_24)

# Concatenar todos os DataFrames em um único
df_21 = pd.concat(dfs_21, ignore_index=True)
df_22 = pd.concat(dfs_22, ignore_index=True)
df_23 = pd.concat(dfs_23, ignore_index=True)
df_24 = pd.concat(dfs_24, ignore_index=True)
```
